```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import math
%matplotlib inline
```

```python
L_wanshen_FMW = pd.read_csv('./FMW/L-wanshen_FMW.csv')
column_L_wanshen_FMW = L_wanshen_FMW.columns
```

```python
data1 = L_wanshen_FMW[column_L_wanshen_FMW[1]][0:2048].reset_index()
data2 = L_wanshen_FMW[column_L_wanshen_FMW[21]][0:2048].reset_index()
data3 = L_wanshen_FMW[column_L_wanshen_FMW[41]][0:2048].reset_index()
data4 = L_wanshen_FMW[column_L_wanshen_FMW[61]][0:2048].reset_index()
data5 = L_wanshen_FMW[column_L_wanshen_FMW[81]][0:2048].reset_index()
data6 = L_wanshen_FMW[column_L_wanshen_FMW[101]][0:2048].reset_index()
```

```python
data1.drop(['index'], axis=1, inplace=True)
data2.drop(['index'], axis=1, inplace=True)
data3.drop(['index'], axis=1, inplace=True)
data4.drop(['index'], axis=1, inplace=True)
data5.drop(['index'], axis=1, inplace=True)
data6.drop(['index'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(25, 8))
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[0]][0:2048], data1)
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[20]][0:2048], data2)
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[40]][0:2048], data3)
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[60]][0:2048], data4)
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[80]][0:2048], data5)
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[100]][0:2048], data6)
```

```python
scaler = StandardScaler()
trans_data1 = scaler.fit_transform(data1)
trans_data2 = scaler.fit_transform(data2)
trans_data3 = scaler.fit_transform(data3)
trans_data4 = scaler.fit_transform(data4)
trans_data5 = scaler.fit_transform(data5)
trans_data6 = scaler.fit_transform(data6)
```

```python
plt.figure(figsize=(30, 8))
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[0]][1024:1536], trans_data1[1024:1536],label='1', color='blue')
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[20]][1024:1536], trans_data2[1024:1536], label='2')
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[40]][1024:1536], trans_data3[1024:1536], label='3')
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[60]][1024:1536], trans_data4[1024:1536], label='4')
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[80]][1024:1536], trans_data5[1024:1536], label='5', color='black')
plt.plot(L_wanshen_FMW[column_L_wanshen_FMW[100]][1024:1536], trans_data6[1024:1536], label='6')
plt.legend()
```
