```python
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.array([0.12,0.21,0.32,0.41,0.51,0.62,0.73,0.86,0.97,0.99])
y = np.array([0.53,0.57,0.63,0.70,0.79,0.91,1.05,1.246,1.42,1.51])
plt.scatter(x,y)
```

```python
class Model(object):
    def __init__(self):
        self.a = tf.Variable(tf.random.uniform([1]))
        self.b = tf.Variable(tf.random.uniform([1]))
        self.c = tf.Variable(tf.random.uniform([1]))
    
    def __call__(self, x):
        return self.a * x * x + self.b * x + self.c
```

```python
def loss_fn(model, x, y):
    y_ = model(x)
    return tf.reduce_mean(tf.square(y_ - y))
```

```python
EPOCHS = 2000
LEARNING_RATE = 0.5

model = Model()

x = tf.constant(x, dtype=tf.float32)
y = tf.constant(y, dtype=tf.float32)

for epoch in range(EPOCHS):
    with tf.GradientTape() as tape:
        loss = loss_fn(model, x, y)
    da, db, dc = tape.gradient(loss,[model.a,model.b,model.c])
    model.a.assign_sub(LEARNING_RATE * da)
    model.b.assign_sub(LEARNING_RATE * db)
    model.c.assign_sub(LEARNING_RATE * dc)
    
    print(f'Epoch [{epoch}/{EPOCHS}], loss [{loss}]')
```

```python
X = tf.linspace(0.0,1.0,50)
plt.scatter(x, y)
plt.plot(X, model(X), c='r')
```

```python
print("{0},{1},{2}".format(model.a,model.b,model.c))
```
