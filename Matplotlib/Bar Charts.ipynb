```python
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
linear_data = np.arange(10)
```

```python
plt.figure()
xvals = list(range(len(linear_data)))
plt.bar(xvals, linear_data, width=0.3)
```

```python
new_xvals = []
for item in xvals:
    new_xvals.append(item+0.3)
```

```python
for x in xvals:
    x = x**2
```

```python
plt.bar(xvals, [x**2 for x in xvals ], width=0.3, color='blue')
```

```python
from random import randint
linear_err = [randint(0,15) for x in range(len(linear_data))]
plt.bar(xvals, linear_data, width=0.3, yerr=linear_err)
```

```python
plt.figure()
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width=0.3, color='b')
plt.bar(xvals, [x**2 for x in linear_data], width=0.3, bottom=linear_data, color='r')
```

```python
plt.figure()
xvals = range(len(linear_data))
plt.barh(xvals, linear_data, height=0.3, color='b')
plt.barh(xvals, [x**2 for x in linear_data], height=0.3, left=linear_data, color='r')
```
