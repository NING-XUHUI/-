```python
for i in energy['Country']:
    pattern = re.compile(r'\D+')
    Country1 = pattern.findall(i)
    newdata =''.join(Country1)
    loc = newdata.find('(')
    if loc >0:
        newdata = newdata[0:loc-1]
    else:
        newdata = newdata
    country.append(newdata.strip())
```
